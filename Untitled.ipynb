```python
def convert_floor(x):
    if isinstance(x, str):
        if 'suterena' in x:
            return -0.5
        if 'poddasze' in x:
            return 1
        if '>' in x:
            return 0
        if 'parter' in x:
            return 0
        if '/' in x:
            parts = x.split('/')
            numerator =  float(parts[0])
            denominator = float(parts[1])
            return numerator / denominator
        else:
            return float(x)
    else:
        return float(x)
```

```python
df_imputed = df2.dropna(subset=['Price','Location','Area','Rooms'], axis=0).copy()
df_imputed['Rent'] = df_imputed['Rent'].fillna('0').str.replace(' zł', '').str.replace(' ', '').str.replace(',', '.').astype(float)
df_imputed['Floor'] = df_imputed['Floor'].apply(convert_floor)
df_imputed['Rent'] = df_imputed['Rent'].fillna(df_imputed['Rent'].mean())
df_imputed['Floor'] = df_imputed["Floor"].fillna(df_imputed['Floor'].mean())
```

```python
X_train,X_test,y_train,y_test = train_test_split(X.to_numpy(),y,test_size=0.2,random_state = 10)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit (n_splits = 50, test_size = 0.2, random_state = 0)

cross_val_score(LinearRegression(),X,y,cv=cv)
```

```python
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression':{
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True,False]
            }
        },
        'lasso':{
            'model': Lasso(max_iter=58000, tol=0.001),
            'params':{
                'alpha' :[1,2],
                'selection':['random','cyclic']
            }
        },
        'decision_tree':{
            'model': DecisionTreeRegressor(),
            'params':
            {
                'criterion' : ['squared_error','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []

    cv = ShuffleSplit(n_splits = 5, test_size = 0.2, random_state = 0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score = False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_})
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

```

```python
find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_price(location,Area,Rooms,Floor,Rent,Building_Ownership,Construction_Status,Outdoor,Heating,Car):
    loc_index = np.where(X.columns == location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = Area
    x[1] = Rooms
    x[2] = Floor
    x[3] = Rent
    x[4] = Building_Ownership
    x[5] = Construction_Status
    x[6] = Outdoor
    x[7] = Heating
    x[8] = Car
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```
